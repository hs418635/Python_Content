For Replacing $ and B with 0000000000

For Replacing $ and B with 000000000 and M with 000000

```python
l=[]
for i in df["Funding"]:
    if "B" in i:
        l.append(int(i[1:-1]+"000000000"))
    elif "M" in i:
        l.append(int(i[1:-1]+"000000"))
    else:
        l.append(None)
```

```python
l
```

```python
for i in df["Funding"]:
    if "M" not in i and "B" not in i:
        print(i)
```

For finding ROI

```python
l=[]
for i,j in zip(df["Valuation"],df["Funding"]):
    l.append(i-j)
```

```python
df["ROI"]=l
```

```python
df
```

```python
df1=df.sort_values("ROI",ascending=False)
```

```python
plt.bar(df1["Company"][:10],df1["ROI"][:10])
f={"size":15,"color":"blue","family":"Times New Roman"}
f1={"size":20,"color":"black","family":"Times New Roman"}
plt.xlabel("Unicorn Name",fontdict=f)
plt.ylabel("Return on Investment",fontdict=f)
plt.title("Unicorn Companies having biggest return",fontdict=f1)
plt.xticks(rotation=60)
plt.rcParams["figure.figsize"]=(7,5)
plt.show()
```

```python
l=[]
for i,j in zip(df["Date Joined"],df["Year Founded"]):
    l.append(int(i[:4])-j)
```

```python
df["Year Taken"]=l
```

```python
df1=df.sort_values("Year Taken")
```

```python
df1.head(10)
```

```python
plt.bar(df1["Company"][1:11],df1["Year Taken"][1:11])
f={"size":15,"color":"blue","family":"Times New Roman"}
f1={"size":20,"color":"black","family":"Times New Roman"}
plt.xlabel("Unicorn Name",fontdict=f)
plt.ylabel("Year Taken to become Unicorn",fontdict=f)
plt.title("Fastest Unicorns",fontdict=f1)
plt.xticks(rotation=60)
plt.rcParams["figure.figsize"]=(7,5)
plt.show()
```

```python
plt.bar(df1["Company"].tail(10),df1["Year Taken"].tail(10))
f={"size":15,"color":"blue","family":"Times New Roman"}
f1={"size":20,"color":"black","family":"Times New Roman"}
plt.xlabel("Unicorn Name",fontdict=f)
plt.ylabel("Year Taken to become Unicorn",fontdict=f)
plt.title("Slowest Unicorns",fontdict=f1)
plt.xticks(rotation=60)
plt.rcParams["figure.figsize"]=(7,5)
plt.show()
```

```python
l=df["Country"].unique()
```

```python
l
```

```python
l1=[]
for i in l:
    l1.append(df["Company"].where(df["Country"]==i).count())
```

```python
df1=pd.DataFrame({"Country":l,"No. of Unicorns":l1})
df1=df1.sort_values("No. of Unicorns",ascending=False)
df1
```

```python
plt.bar(df1["Country"][:10],df1["No. of Unicorns"][:10])
f={"size":15,"color":"blue","family":"Times New Roman"}
f1={"size":20,"color":"black","family":"Times New Roman"}
plt.xlabel("Country",fontdict=f)
plt.ylabel("No. of Unicorns",fontdict=f)
plt.title("Country Wise No. of Unicorns",fontdict=f1)
plt.xticks(rotation=60)
plt.rcParams["figure.figsize"]=(7,5)
plt.show()
```
