### Method vs Functions
1) We always define a method inside a class.
2) A method definition always includes 'self' or other word as its parameter but in funcion there is no need to pass the parameter.
3) And lastly, we call a method on an object using the dot operator.

```python
# function
def message():
    print("India is a greate country")
message()
```

```python
# function
def add(a,b):
    return a + b
add(5,6)
```

```python
# method
class number:
    def multi(x,y):
        print(x*y)
multi()
```

```python
class number:
    def multi(x,y): # its a method so we have  to pass self argument , which belongs to object
        print(x*y)
obj = number()
obj.multi(5,4)
```

```python
class number:
    def multi(self,x,y): # this self belongs to obj we have made
        print(x*y)
obj = number()
obj.multi(5,4)
```

Instance variable is the variable which is defined inside the function and call be called inside function only

```python
class test():
    def demo():
        a=100 #esme function ke andr variable defined hai esliye sirf function me hi call kiya ja skta hai
        print("instance variable is ",a)
test.demo()
print("clas variable is ",a)#is wale line me error dega bcz instance variable ko bahar call ni kr skte
# esme variable function ke andr defined hai esliye last wali print call me error dega name a is not defined 
```

#### Constructor

__init__  they are inbuilt and special function, there is no need to call them, it initializes the variable

```python
class value:
    def __init__(self,a,b):
        self.a=a
        self.b=b
        print("constructor is called")
    def display(self):
        print("initialized value are:",self.a,self.b)
v=value(1,2)
v.display()
t=value(6,7)
t.display()
```

#### Destructor

Deleting in python automatically

```python
class value:
    def __init__(self,a,b):
        self.a=a
        self.b=b
        print('constructor is called')
    def display(self):
        print('initialized value are:',self.a,self.b)
    def __del__(self):
        print('Destructor is called')
v=value(2,3)
v.display()
t=value(6,7)
t.display()
```

#### Now we will delete the values by using destructor

```python
class value:
    def __init__(self,a,b):
        self.a=a
        self.b=b
        print('constructor is called')
    def display(self):
        print('initialized value are:',self.a,self.b)
    def __del__(self):
        print('Destructor is called')
v=value(2,3)
v.display()
t=value(6,7)
t.display()
del v #delete function ko call krna padta hai
del t
v
t #after deleting it will through error that v is not defined
```

### Encapsulation 

Wrapping of data

```python
class Employee:
    #Common base class for all employee
    empCount=0 #how many employee
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
        Employee.empCount +=1
        print('Constructor is called')
    
    def displayEmployee(self):
        print('Name',self.name,'salary:-',self.salary)
        
    def __del__(self):
        print('Destructor is called')
        
emp1=Employee('Rohit',30000)
emp1.displayEmployee()
emp2=Employee('Rahul',50000)
emp2.displayEmployee()
emp3=Employee('shubham',40000)
emp3.displayEmployee()

print("Total Employee %d" %Employee.empCount)
```

#### Now calling destructor

```python
class Employee:
    #Common base class for all employee
    empCount=0 #how many employee
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
        Employee.empCount +=1
        print('Constructor is called')
    
    def displayEmployee(self):
        print('Name',self.name,'salary:-',self.salary)
        
    def __del__(self):
        print('Destructor is called')
        
emp1=Employee('Rohit',30000)
emp1.displayEmployee()
emp2=Employee('Rahul',50000)
emp2.displayEmployee()
emp3=Employee('shubham',40000)
emp3.displayEmployee()

print("Total Employee %d" %Employee.empCount)

del emp1
del emp2
del emp3
```

#### SIngle level inheritance

```python
# single inheritance: ek object se aap dono method ko call kr rhe ho 
class A:#making class A
    def displayA(self):#creating funtion
        print("welcome to dalal street A")
        
class B(A):#b me a ko inherit krwa diya
    def displayB(self):
        print("welcome to dalal street B")
        
obj=B()#making object of B class
obj.displayA()#calling class A
obj.displayB()#calling Class B, dono methods call ho rhe hai ek object se
```

#### Multiple Inheritance

```python
# Multiple inheritance
class A:
    def show(self):
        print('this is class A')
class B(A):
    def show1(self):
        print('this is class B')
class C(B,A):
    def show2(self):
        print('this is class C')
class D(C,B,A):
    def show3(self):
        print('this is class D')
        
k=D()
k.show2()
k.show()
k.show1()
k.show3()
```

#### Hierrachical inheritance

```python
# Hierrachical inheritance

class A: #father
    def show(self):
        print('this is class A of father ')
class B(A):#son
    def show1(self):
        print('this is class B of son')
class C(A):#daughter
    def show2(self):
        print('this is class C of daughter')
        
s=B() #son class se object bna rhe
s.show1()
s.show()
# s.show2() B object has no attribute show2 error ayegi
```

```python
# Hierrachical inheritance

class A: #father
    def show(self):
        print('this is class A of father ')
class B(A):#son
    def show1(self):
        print('this is class B of son')
class C(A):#daughter
    def show2(self):
        print('this is class C of daughter')
        
s=C()
# s.show1() ye show1 attribute ni hai C class ke andr error dega ye
s.show()
s.show2()
```

#### Multi level inheritance

```python
# multi level inheritance

class A:
    def show(self):
        print('this is class A')
class B(A):
    def show1(self):
        print('this is class B')
class C(B):
    def show2(self):
        print('this is class C')
class D(C):
    def show3(self):
        print('this is class D')
        
j=D()
j.show()
j.show2()
j.show1()
j.show3()
```

### Polymorphism

Polymorphism in Python means same function name being used for different types. In polymorphism, arguments and list length can be counted. Overloading and overriding concepts are also discussed with the process of use of these function.

When child function overrides the parents function, the function which is used for calling the parents function back is super.

```python
#polymorphism:function overloading and function overwriting
l=[10,20,30,40]
print(len(l))
s="himanshu"
print(len(s))#function same hua aur andr ka signature/datatype change
#ho gya ek bar l ko count kiya aur ek bar s ko.. esi process ko ham
#polymorphism kahte hai
```

#### Overloading

```python
# overloading;fun ka name same hoga but parameter apne according change kr diya
class Ws:
    def displayinfo(self,name=''):
        print("welcome to dalal street" + name)
        
obj=Ws()
obj.displayinfo()
#abhi maine parameter pass nhi kiya       
#jabki name vaha pe liya hai.. so this overloading ka concept
#concept- function ka naam to same hai but apne parameter pass nahi kiya
```

#### Overwirting

```python
# overwriting:function ko overwrite karana
class ws:
    def displayinfo(self): #fun1 same
        print("welcom to dalal street")
        
class IIP(ws):#fun2 same,inherit kiya ws ko
    def displayinfo(self):
        print("welcom to IIP")
        
obj=IIP()
obj.displayinfo() #jab inherit kiya to overwrite ho gya
```

#### Super() 
super function ki help se parent ke function ko call kr skte hai jo same name ka hai

```python
# overriding:function ko overriding karana
class ws:
    def displayinfo(self): #fun1 same
        print("welcom to dalal street")
        
class IIP(ws):#fun2 same,inherit kiya ws ko
    def displayinfo(self):
        super().displayinfo()
        print("welcom to IIP")
        
obj=IIP()
obj.displayinfo() #jab inherit kiya to overwrite ho gya
```

### Abstraction in OOP's

```python
#data abstraction #OOPs
from abc import ABC, abstractmethod

class parent(ABC):
    
    @abstractmethod
    def fun1(self):
        pass
    def fun2(self):
        print("this is an parent class")
        
class child(parent):
    def fun1(self):
        print("this is an child class")
        
obj=child()
obj.fun1()
obj.fun2()
```

#### Bank Account

```python
class Account():
    
    def __init__(self,owner,balance=0):
        
        self.owner = owner
        self.balance = balance
        
    def deposit(self,dep_amt):
        
        self.balance = self.balance + dep_amt
        print(f"Added {dep_amt} to the balance")
        
    def withdrawal(self,wd_amt):
        
        if self.balance >= wd_amt:
            self.balance = self.balance - wd_amt
            print("withdrawal accepted")
        else:
            print("Sorry non enough funds! ")
            
    def __str__(self):
        return "Owner : {self.owner} \nBalance: {self.balance}"
```
