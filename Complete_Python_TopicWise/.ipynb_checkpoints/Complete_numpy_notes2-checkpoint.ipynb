# iteration over array

Iterating means going through elements one by one

As we deal with multi-dimensional arrays , we can do this using basic for loops of python

if we iterate on a 1D array it will go through each element one by one

### 1D array

```python
import numpy as np
arr = np.arange(4)
print(arr)
print("dimension:",arr.ndim)
```

```python
for i in arr:
    print(i)
```

```python
arr
```

### get 2 from this 1d array using loop

```python
for i in range(len(arr)):
    if i==2:
        output=arr[i]
        break
output
```

### 2D array

In a 2D array it will go through all the rows

```python
np.random.seed(1)
arr = np.random.randint(1,6,6).reshape(2,3)
print(arr)
print('dim:',arr.ndim)
```

```python
for i in arr:
    print(i)
```

```python
# iterate on each scalar element of the 2D array
for i in arr:
    for j in i:
        print(j)
```

### extract 4 from 1st indexed array

```python
arr
```

```python
for i in arr[1]:
    if i==4:
        output = i
        break
output
```

### 3D array

In a 3D array it will go through all the 2D arrays

#### To return the actual values, the scalars , we have to iterate the arrays in each dim.

```python
np.random.seed(123)
arr3d = np.random.randint(1,18,18).reshape(3,2,3)
print(arr3d)
print('dim:',arr3d.ndim)
```

```python
# accessing 2D matrices
for i in arr3d:
    print(i) 
```

```python
# accessing rows of 2D matrices
for i in arr3d:
    for j in i:
        print(j)
```

```python
# accessing each values/scalars
for i in arr3d:
    for j in i:
        for k in j:
            print(k)
```

### extract 17 from 2nd indexed 2d array inside 3d array 

```python
arr3d
```

```python
for i in arr3d[2]:
    for j in i:
        if j == 17:
            output = j
            break
output
```

### Iterating arrays using nditer()

using nested loops will not be good for memory and Run time , so to replace nested loops 

we will use nditer()

#### nditre() --> can be used from very basic to very advanced iterations. it solves some basic issues which me face in iteration

```python
print(arr3d)
```

```python
# using nditer()
for i in np.nditer(arr3d):
    print(i)
```

## Iterating arrays with defferent data types

We can use op_dtypes argument and pass it to the expected data type to change the datatype of element while iterating.

example: for i in np.nditer(arr,flags=['buffered'],op_dtype=['S'])
    
NumPy does not change the data type of the element in-place (where the element is in array) so it needs some other space to perform this action, that extra space is called buffer, and in order to enable it in nditer() we pass flags=['buffered'].

```python
arr = np.array([1,2,3,4])
for i in np.nditer(arr,flags=['buffered'],op_dtypes=["S"]):
    print(i)
```

### Iterating with different step size

we can use filtering and followed by iteration

```python
arr = np.arange(10).reshape(2,5)
print(arr)
print('dim',arr.ndim)
```

```python
for i in np.nditer(arr[:,::2]):
    print(i)
```

### Dot(Scalar) Product
The dot product or scalar product is an algebraic operation that takes two equal-length sequences of numbers (usually coordinate vectors), and returns a single number.

```python
arr1 = np.array([[3,5],[9,4]]) 
arr2 = np.array([[11,1],[23,12]]) 
arr1
```

```python
arr2
```

### Axes in numpy

```python
# axes 0 ----> rows
# axes 1 ----> column

arr = np.arange(1,7).reshape(2,3)
arr
```

```python
print("cutting the array vertically ")
print(np.sum(arr,axis=0))
```

```python
print("cutting the array Horizontally ")
print(np.sum(arr,axis=1))
```
