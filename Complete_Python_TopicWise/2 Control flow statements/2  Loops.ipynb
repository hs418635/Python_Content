```python
# print stars in right angle triangle shape
def print_star_triangle(rows):
    i = 1
    while i <= rows:
        j = 1
        while j <= i:
            print("*", end=" ")
            j += 1
        print()
        i += 1

# Example usage
print_star_triangle(6)
```

```python
rows = int(input("Enter a number: "))
i = rows

while i >= 1:
    j = 1
    # Print spaces
    while j <= rows - i:
        print(" ", end=" ")
        j += 1
    # Print stars
    j = 1
    while j <= i:
        print("*", end=" ")
        j += 1
    print()
    i -= 1
```

```python
k=1
i=1
rows=int(input("enter number of rows to print: "))
while i <= rows:
    # printing space
    s = 1 # to print space 
    while s <= rows - i:
        print(" ", end=" ") # " " used to print space , end=" " used to stay the cursor on the same line 
        s = s + 1
    # printing start 
    j = 1
    while j <= k:
        print("*", end=" ")
        j = j + 1
    
    k = k + 2 # increasing k with 2 because in the second line printing 3 stars , 5 on third , cycle goes on
    print() # to shift the cursor on the next line
    i=i+1      
```

```python
# Write a program that prints the nth term of  Fibonacci series.
num = int(input('Enter the term number: '))

# Initial terms of the Fibonacci sequence
n1, n2 = 0, 1

# Counter variable to keep track of current term number
count = 0

# Check if the input is valid
if num <= 0:
    print("Please enter a positive integer.")
elif num == 1:
    print("Fibonacci sequence upto", num, ":")
    print(n1)
else:
    while count < num:
        nth_term = n1
        nth_sum = n1 + n2
        # update values for next iteration
        n1 = n2
        n2 = nth_sum
        count += 1

    print("The", num, "th Fibonacci term is:", nth_term)
```

### Nested For loop

```python
# Write a program that takes two numbers as input and prints all the prime numbers lying between them.
N1=int(input('enter first number : '))
N2=int(input('enter second number : '))
print('Prime numbers between',N1,'and',N2,'are: ')
for i in range(N1,N2+1):
    if i > 1:
        for j in range(2,i):
            if (i % j) == 0:
                break # avoid internal else 
        else:
            print(i, end=' ')
```

```python
# Q1.Write a Python program using a while loop that prints numbers from 1 to 10. Stop the loop when the number is 5.
i = 1
while i <= 10:
    if i == 5:
        break
    print(i,end=" ")
    i += 1
```

```python
# Q2.Write a Python program using a for loop that iterates over a string "Elephant" and prints each letter. Stop the loop if the letter is "h".
s = "Elephant"
for i in s:
    if i == "h":
        break
    print(i, end="")
```

#### Continue

```python
# Q3.Write a Python program using a while loop that prints numbers from 1 to 10. Skip the number 5.
i = 1
while i <= 10:
    if i == 5:
        i += 1
        continue
    print(i, end=" ")
    i += 1
```

```python
# Q4.Write a Python program using a for loop that iterates over a range from 1 to 10 and prints each number. Skip the number 7.
for i in range(1,11):
    if i==7:
        continue
    print(i,end=" ")
```
