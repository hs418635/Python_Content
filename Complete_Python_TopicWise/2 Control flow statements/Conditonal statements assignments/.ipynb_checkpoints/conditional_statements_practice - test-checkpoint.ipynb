# Conditional Statements Practice Questions and Answers

## If-Else Practice Questions

```

# Odd or Even
num = 27
if num % 2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")  # Output: 27 is odd
```

```

# Positive or Negative
num = -45
if num >= 0:
    print(f"{num} is positive")
else:
    print(f"{num} is negative")  # Output: -45 is negative
```

```

# Greater Number
num1 = 34
num2 = 29
if num1 > num2:
    print(f"{num1} is greater than {num2}")  # Output: 34 is greater than 29
else:
    print(f"{num2} is greater than {num1}")
```

```

# Check Divisibility
num = 48
if num % 6 == 0:
    print(f"{num} is divisible by 6")  # Output: 48 is divisible by 6
else:
    print(f"{num} is not divisible by 6")
```

```

# Vowel or Consonant
char = 'e'
if char in 'aeiou':
    print(f"{char} is a vowel")  # Output: e is a vowel
else:
    print(f"{char} is a consonant")
```

## If-Elif-Else Practice Questions

```

# Grade Calculation
score = 85
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")  # Output: Grade: B
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```

```

# Day of the Week
day_number = 4
if day_number == 1:
    print("Monday")
elif day_number == 2:
    print("Tuesday")
elif day_number == 3:
    print("Wednesday")
elif day_number == 4:
    print("Thursday")  # Output: Thursday
elif day_number == 5:
    print("Friday")
elif day_number == 6:
    print("Saturday")
elif day_number == 7:
    print("Sunday")
else:
    print("Invalid day number")
```

```

# Temperature Check
temperature = 30
if temperature < 15:
    print("Cold")
elif temperature <= 25:
    print("Warm")
else:
    print("Hot")  # Output: Hot
```

```

# Age Group
age = 23
if age <= 12:
    print("Child")
elif age <= 19:
    print("Teen")
elif age <= 59:
    print("Adult")  # Output: Adult
else:
    print("Senior")
```

```

# Traffic Light
light_color = 'red'
if light_color == 'red':
    print("Stop")  # Output: Stop
elif light_color == 'yellow':
    print("Slow down")
elif light_color == 'green':
    print("Go")
else:
    print("Invalid color")
```

## Nested If Practice Questions

```

# Leap Year
year = 2000
if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print(f"{year} is a leap year")  # Output: 2000 is a leap year
        else:
            print(f"{year} is not a leap year")
    else:
        print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```

```

# Number Classification
num = -3
if num > 0:
    if num % 2 == 0:
        print(f"{num} is positive and even")
    else:
        print(f"{num} is positive and odd")
elif num < 0:
    if num % 2 == 0:
        print(f"{num} is negative and even")
    else:
        print(f"{num} is negative and odd")  # Output: -3 is negative and odd
else:
        print(f"{num} is zero")
```

```

# Triangle Type
a = 3
b = 4
c = 5
if a == b == c:
    print("Equilateral triangle")
elif a == b or b == c or a == c:
    print("Isosceles triangle")
else:
    print("Scalene triangle")  # Output: Scalene triangle
```

```

# Voting Eligibility
age = 20
if age >= 18:
    print("Eligible to vote")  # Output: Eligible to vote
    if age >= 60:
        print("Eligible for senior citizen discount")
    else:
        print("Not eligible for senior citizen discount")
else:
    print("Not eligible to vote")
```

```

# Nested If Example
num = 29
if num > 1:
    for i in range(2, num):
        if (num % i) == 0:
            print(f"{num} is not a prime number")
            break
    else:
        print(f"{num} is a prime number")  # Output: 29 is a prime number
else:
    print(f"{num} is not a prime number")
```
