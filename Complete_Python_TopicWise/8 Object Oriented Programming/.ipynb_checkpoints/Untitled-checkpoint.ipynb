```python
# method vs functions

def message():
    print("Hi there is monday.")
message()
```

```python
def add(a,b):
    print(a+b)
add(2,3)
```

```python
class number:
    def multi(x,y):
        return x*y
multi(2,3)
```

```python
class number: # class
    def multi(x,y): # method
        return x*y
number.multi(2,3)
```

```python
# classes , objects and attributes

# classes --> are the blueprint for creating the objects . They encapsulate data for the objects and methods to manipulate that data.

# objects --> are the real world entity that has state and behaviour associated with it

# attributes --> are the variable that belong to an object or class
```

```python
# classess and objects 

class sample:
    def show():
        print("hello world")
sample.show()
```

```python
class sample:
    def show(self):
        print("hello world")
obj = sample() # object / instance / instances of the class
obj.show()
```

```python
# class variable --> are the variable that are defined inside the class but outside any method
# and can be shared accross all instances of the class.

class test():
    a = "i am a class variable"
    def show(self):
        print("class variable inside the method is  :",a)
obj = test()
obj.show()
```

```python
class test():
    a = "i am a class variable"
    def show(self):
        print("class variable inside the method is  :",test.a)
obj = test()
obj.show()
print("class variable outside the method is  :",test.a)
```

```python
# instance variable
class test():
    def show(self):
        a = "i am a class variable"
        print("instance variable inside the method is  :",a)
obj = test()
obj.show()
print("instance variable outside the method is  :",a)
```

```python
class Test:
    a = "i am class variable"
    def demo(self):
        b = "i am instance variable"
        print("instance variabel is :",b)
        print("instance variable inside the method is :",Test.a)
obj = Test()
obj.a
obj.demo()
print("class variabel is :",obj.a)
```
