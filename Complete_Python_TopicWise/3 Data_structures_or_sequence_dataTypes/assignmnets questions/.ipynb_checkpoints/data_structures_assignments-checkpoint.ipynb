### string assignment questions

```python
# 1)  Calculate the Length of a String
#     Write a Python program to calculate the length of a given string.
#     print the result using string formatting

string = "Hello, World!"
length = len(string)
print(f"The length of the string '{string}' is {length}.")
```

```python
# 2) Count Character Frequency
#    Write a Python program to count the number of characters (character frequency) in a given 
#    string. For example, for the string "google.com", 
#    the expected output would be {'g': 2, 'o': 3, 'l': 1, 'e': 1, '.': 1, 'c': 1, 'm': 1}.


string = "google.com"
frequency = {}
for char in string:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1
print(frequency)
```

```python
# 3) First and Last Two Characters
#    Write a Python program to get a string made of the first 2 and last 2 characters of a given string. If the string length is less than 2, return the empty string instead. For example, for the string "w3resource",
#    the expected output would be "w3ce".
string = "w3resource"
first_two = string[:2]
last_two = string[-2:]
result = first_two + last_two
print(result)
```

```python
# 4) Swap First Two Characters of Each String
#    Write a Python program to get a single string from two given strings, separated by a space, and swap the first two characters of each string. For example, for the strings "abc" and "xyz", 
#    the expected output would be "xyc abz".

string1 = "abc"
string2 = "xyz"
result = string2[:2] + string1[2:] + " " + string1[:2] + string2[2:]
print(result)
```

```python
# using loop
x = "abc"
y = "xyz"
r1 = ''
r2 = ''
    
# Swap the first two characters of each string
for i in range(len(x)):
    if i < 2:
        r1 += x[i]
    else:
        r1 += y[i]
    
for j in range(len(y)):
    if j < 2:
        r2 += y[j]
    else:
        r2 += x[j]
    
print(r2 + ' ' + r1)
```

```python
# 5) Remove Repeated Consecutive Characters
#    Write a Python program to remove repeated consecutive characters and replace them with single letters. For example, for the string "aabbbcdeffff",
#    the expected output would be "abcdef".

string = "aabbbcdeffff"
result = ""
prev_char = ""
for char in string:
    if char!= prev_char:
        result += char
    prev_char = char
print(result)  
```

```python
# 6) Count Same Three Letters at Same Index
#    Write a Python program that takes two strings and counts the number of times each string contains the same three letters at the same index. 
#    For example, for the strings "Red RedGreen" and "Red White Red White", the expected output would be 1 and 7, respectively.

string1 = "Red RedGreen"
string2 = "Red White Red White"

# Initialize counts
count1 = 0
count2 = 0

# Count occurrences of substrings of length 3 that match in both strings
for i in range(len(string1) - 2):  # -2 to ensure we don't go out of bounds
    if string1[i:i+3] == string2[i:i+3]:
        count1 += 1

for i in range(len(string2) - 2):  # -2 to ensure we don't go out of bounds
    if string2[i:i+3] == string1[i:i+3]:
        count2 += 1

print(count1, count2)
```

```python
# 7) Insert Space Before Capital Letters
#    Write a Python program that takes a string and inserts a space before every capital letter appears in a given word. 
#    For example, for the string "PythonExercises", the expected output would be "Python Exercises".

string = "PythonExercises"
result = ""
for char in string:
    if char.isupper():
        result += " " + char.lower()
    else:
        result += char
print(result)
```

```python
# 8) Replace Characters with Their Numbers
#    Write a Python program that takes a string and replaces all the characters with their respective numbers.
#    For example, for the string "Python", the expected output would be "16 25 20 8 15 14".

string = "Python"
numbers = {
    'P': '16',
    'y': '25',
    't': '20',
    'h': '8',
    'o': '15',
    'n': '14'
}
result = " ".join(numbers[char] for char in string)
print(result)
```

### List assignment questions

```python
# 1) Sum all the items in a list: 
#    Write a program that takes a list of numbers as input and prints the sum of all the items in the list.
numbers = [1, 2, 3, 4, 5]
print("Sum:", sum(numbers))
```

```python
# 2) Multiply all the items in a list: 
#    Write a program that takes a list of numbers as input and prints the product of all the items in the list.
product = 1
for num in numbers:
    product *= num
print("Product:", product)
```

```python
# 3) Get the largest number from a list: 
#    Write a program that takes a list of numbers as input and prints the largest number in the list.
largest = max(numbers)
print("Largest:", largest)
```

```python
# 4) Get the smallest number from a list: 
#    Write a program that takes a list of numbers as input and prints the smallest number in the list.
smallest = min(numbers)
print("Smallest:", smallest)
```

```python
# 5) Count the number of strings from a given list of strings: 
#    Write a program that takes a list of strings as input and counts the number of strings where the string length is 2 or more and the first and last characters are the same.
strings = ['abc', 'xyz', 'aba', '1221']
count = 0
for s in strings:
    if len(s) >= 2 and s[0] == s[-1]:
        count += 1
print("Count of matching strings:", count)
```

```python
# 6) Get a list, sorted in increasing order by the last element in each tuple: 
#    Write a program that takes a list of non-empty tuples as input and prints the list sorted in increasing order by the last element in each tuple.
tuples = [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]
sorted_tuples = sorted(tuples, key=lambda x: x[-1])
print("Sorted by last element:", sorted_tuples)
```

```python
# 7) Remove duplicates from a list: 
#    Write a program that takes a list as input and prints the list with duplicates removed.
unique_numbers = list(set(numbers))
print("Unique numbers:", unique_numbers)
```

```python
# 8) Check if a list is empty or not: 
#    Write a program that takes a list as input and checks if the list is empty or not, printing "Empty" if true and "Not Empty" if false.
if not numbers:
    print("Empty")
else:
    print("Not Empty")
```

### Set assignment question

```python
# Write a Python program that performs the following operations on sets:

# Addition: Create two sets, setA and setB, and add all elements from setB into setA.
# Intersection: Find the intersection of setA and setB, meaning the elements that are common to both sets.
# Difference: Find the difference between setA and setB, meaning the elements that are in setA but not in setB.
# Symmetric Difference: Find the symmetric difference between setA and setB, meaning the elements that are in either of the sets but not in both.
# Union: Find the union of setA and setB, meaning all elements that are in either of the sets.
```

```python
# Step 1: Addition
setA = {"apple", "banana", "cherry"}
setB = {"banana", "cherry", "date", "elderberry"}

# Adding all elements from setB into setA
setA.update(setB)
print("After addition:", setA)

# Step 2: Intersection
intersection = setA & setB
print("Intersection:", intersection)

# Step 3: Difference
difference = setA - setB
print("Difference:", difference)

# Step 4: Symmetric Difference
symmetric_difference = setA ^ setB
print("Symmetric Difference:", symmetric_difference)

# Step 5: Union
union = setA | setB
print("Union:", union)

```

### Dictionary

```python
# Create a dictionary named student_grades that stores the names of students as keys and their grades as values. The grades should be stored as a list of integers representing the scores for each subject (e.g., Math, English, Science). After creating the dictionary, perform the following operations:

# Print the dictionary.
# Add a new student named "John" with grades [85, 90, 88].
# Remove the student "John" from the dictionary.
# Check if the student "John" exists in the dictionary.
# Print the average grade for each subject across all students.
# Find the student with the highest total grade and print their name and total grade.
```

```python
# Step 1: Create the dictionary
student_grades = {
    "Alice": [95, 92, 88],
    "Bob": [78, 85, 90],
    "Charlie": [92, 88, 95]
}

# Step 2: Add a new student
student_grades["John"] = [85, 90, 88]

# Step 3: Remove the student
del student_grades["John"]

# Step 4: Check if the student exists
if "John" in student_grades:
    print("John exists in the dictionary.")
else:
    print("John does not exist in the dictionary.")

# Step 5: Calculate and print the average grade for each subject
subject_averages = {}
for student, grades in student_grades.items():
    for i, grade in enumerate(grades):
        if i not in subject_averages:
            subject_averages[i] = []
        subject_averages[i].append(grade)
    # Calculate the average for each subject
    for subject, grades in subject_averages.items():
        average_grade = sum(grades) / len(grades)
        print(f"Average grade for subject {subject + 1}: {average_grade}")

# Step 6: Find the student with the highest total grade
highest_total_grade = 0
highest_total_student = ""
for student, grades in student_grades.items():
    total_grade = sum(grades)
    if total_grade > highest_total_grade:
        highest_total_grade = total_grade
        highest_total_student = student

print(f"The student with the highest total grade is {highest_total_student} with a total grade of {highest_total_grade}.")
```
