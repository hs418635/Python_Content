##### what is file handling ?
File handling means :
1) opening a file.
2) Performing some operations on it.
3) closing a file.

File Handling Modes--> mode specifies the purpose of opening file
1) text mode--> In this mode when you open the file , python treats its content as str type.

   When you get the data from a text mode file , python first decodes the raw bytes using either a platform dependent encoding or specified encoding
   


2) binary mode --> In the binary mode python uses the data without any encoding because its content is treated as bytes . Eg. images , audios , vedios etc

##### way of closing files
1) Normal way
2) using exception handling
3) with statement

```python
# normal way
f = open("about_india.txt",'r')
print(f.read())
f.close()

# Exceptions stops normal flow of programm--> suppose while doing operation you got exception, you will get exception error
```

```python
# # using exception handling

try:
    f=open("about_india.txt",mode='r')
    print(f.read())
finally:             # finally block always gets executed whether exception occurs or not
    f.close()
```

```python
# using with statement

with open("about_india.txt",mode='r') as f:
    data=f.read()
    print(data)
# there is no need to close using this method python automatically closes the file
```

##### important method
1) tell() method --> is used to find the current position of a file pointer/curson from the begining of the file, position starts from 0 , it is just like indexing in string.
2) seek() --> is used to change the position of file pointer.

```python
# tell() method
f = open("about_india.txt",'r') 
print(f.tell()) # we have not read any character yet , so it will return 0th position
f.close()
```

```python
# tell() method
f = open("about_india.txt",'r') 
print(f.tell())
f.read(3)
print(f.tell()) # after reading 3 characters , the cursor position will be 3
f.read() # reading all character
print(f.tell()) # now the cursor will be at last
f.close()
```

```python
# seek()
f = open("about_india.txt",'r') 
print(f.tell()) # 0
print(f.read(3)) # here we have read 3 character 
f.seek(0) # now again i want to read from starting
print(f.read())
f.close()
```

##### Finding number of characters , words and lines in file

```python
f = open("about_india.txt",'r') 
count_words = 0
count_chars = 0
count_lines = 0
for line in f:
    count_lines += 1
    lines = line.strip("\n")
    # print("lines :",lines)
    count_chars  += len(lines)
    list1 = lines.split()
    # print("list1 :",list1)
    count_words += len(list1)
f.close()
print('No of lines :',count_lines)
print('No of words :',count_words)
print('No of characters :',count_chars)
```

### 4.r+ modes and w+ modes

r+ mode
1) read and write mode
2) the file pointer position at the begining of the file
3) the r+ throws an error if file does not exist
4) opens an existing file without truncating it
5) Note :- use when you want to read data first and then write(append) data to same file.

w+ mode
1) write and read mode.
2) the file pointer position at the beginint of the file.
3) the w+ creates a new file if file does not exist.
4) opens an existing file and truncate it first.

### 5.merge different files into one

```python
import os
files = []
merged_data = ""
while True:
    f_name = input("Enter the file name :")
    files.append(f_name)
    ans = input("Want to add another file ? (y/n):").lower()
    if ans != 'y':
        break

for file in files:
    filename = file + '.txt'
    if os.path.isfile(filename):
        f = open(filename,mode='r',encoding='utf-8')
        merged_data = merged_data + f.read() + '\n'
        f.close()
print(merged_data)

with open(input("Enter the final file") + '.txt', mode='x',encoding='utf-8') as f:
    f.write(merged_data)
    print("merging is done, thanks !!")
```
