```python
import nbformat as nbf

# List of questions and answers
qa_pairs = [
    ("Reading a File:\nWrite a Python program to read the content of a text file named `example.txt` and display it.",
    """# Answer
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
    """),

    ("Writing to a File:\nWrite a Python program to write \"Hello, World!\" to a new file named `hello.txt`.",
    """# Answer
with open('hello.txt', 'w') as file:
    file.write("Hello, World!")
    """),

    ("Appending to a File:\nWrite a Python program to append the line \"This is an appended line.\" to the file `hello.txt`.",
    """# Answer
with open('hello.txt', 'a') as file:
    file.write("\\nThis is an appended line.")
    """),

    ("Copying a File:\nWrite a Python program to copy the content of `example.txt` to a new file named `example_copy.txt`.",
    """# Answer
with open('example.txt', 'r') as source_file:
    content = source_file.read()

with open('example_copy.txt', 'w') as dest_file:
    dest_file.write(content)
    """),

    ("Reading Specific Lines:\nWrite a Python program to read the first 5 lines of a file named `example.txt`.",
    """# Answer
with open('example.txt', 'r') as file:
    for _ in range(5):
        line = file.readline()
        if not line:
            break
        print(line.strip())
    """),

    ("Counting Lines:\nWrite a Python program to count the number of lines in a file named `example.txt`.",
    """# Answer
with open('example.txt', 'r') as file:
    line_count = sum(1 for line in file)
print(f"Number of lines: {line_count}")
    """),

    ("Counting Words:\nWrite a Python program to count the number of words in a file named `example.txt`.",
    """# Answer
with open('example.txt', 'r') as file:
    words = file.read().split()
    word_count = len(words)
print(f"Number of words: {word_count}")
    """),

    ("Counting Characters:\nWrite a Python program to count the number of characters in a file named `example.txt`.",
    """# Answer
with open('example.txt', 'r') as file:
    content = file.read()
    char_count = len(content)
print(f"Number of characters: {char_count}")
    """),

    ("Removing Blank Lines:\nWrite a Python program to remove all blank lines from a file named `example.txt`.",
    """# Answer
with open('example.txt', 'r') as file:
    lines = file.readlines()

with open('example.txt', 'w') as file:
    for line in lines:
        if line.strip():
            file.write(line)
    """),

    ("Reading a File in Reverse:\nWrite a Python program to read a file named `example.txt` in reverse order (line by line).",
    """# Answer
with open('example.txt', 'r') as file:
    lines = file.readlines()
for line in reversed(lines):
    print(line.strip())
    """),

    ("Finding the Longest Word:\nWrite a Python program to find the longest word in a file named `example.txt`.",
    """# Answer
with open('example.txt', 'r') as file:
    words = file.read().split()
    longest_word = max(words, key=len)
print(f"Longest word: {longest_word}")
    """),

    ("Replacing a Word:\nWrite a Python program to replace the word \"old\" with \"new\" in a file named `example.txt`.",
    """# Answer
with open('example.txt', 'r') as file:
    content = file.read()
content = content.replace('old', 'new')
with open('example.txt', 'w') as file:
    file.write(content)
    """),

    ("Merging Two Files:\nWrite a Python program to merge the content of `file1.txt` and `file2.txt` into a new file named `merged.txt`.",
    """# Answer
with open('file1.txt', 'r') as file1, open('file2.txt', 'r') as file2:
    content1 = file1.read()
    content2 = file2.read()

with open('merged.txt', 'w') as merged_file:
    merged_file.write(content1 + '\\n' + content2)
    """),

    ("Reading a File Line by Line:\nWrite a Python program to read a file line by line and store each line in a list.",
    """# Answer
lines = []
with open('example.txt', 'r') as file:
    for line in file:
        lines.append(line.strip())
print(lines)
    """),

    ("Writing a List to a File:\nWrite a Python program to write a list of strings to a file named `list_output.txt`.",
    """# Answer
lines = ["Line 1", "Line 2", "Line 3"]
with open('list_output.txt', 'w') as file:
    for line in lines:
        file.write(line + '\\n')
    """)
]

# Creating a new Jupyter notebook
nb = nbf.v4.new_notebook()

# Adding the questions and answers as cells
for question, answer in qa_pairs:
    nb.cells.append(nbf.v4.new_markdown_cell(f"**Question:**\n{question}"))
    nb.cells.append(nbf.v4.new_code_cell(answer))

# Saving the notebook to a file
nb_path = "file_handling_practice.ipynb"
with open(nb_path, 'w') as f:
    nbf.write(nb, f)
```
