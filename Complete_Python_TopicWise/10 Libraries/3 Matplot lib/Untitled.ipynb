```python
# loading data
df = sns.load_dataset('tips')
```

```python
# scatterplot
sns.scatterplot(df,x='tip',y='total_bill')
plt.show()
```

```python
# lineplot
sns.lineplot(df,x='size',y='total_bill')
plt.show()
```

```python
# barplot
sns.barplot(df,x='day',y='total_bill')
plt.show()
```

```python
sns.boxplot(df,x='day',y='total_bill')
plt.show()
```

```python
sns.histplot(df)
plt.show()
```
