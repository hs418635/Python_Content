#### Dotted/Dashed lines

```python
# dotted , dashed
x= [1, 2, 3, 4]
y= [10, 20, 25, 30]
plt.plot(x,y,color='green',linestyle="dashed",linewidth=3)
```

#### markers

```python
# v--> downward triangle , * , o-->circle , ^---> for upward triangle , d--diamond
x= [1, 2, 3, 4]
y= [10, 20, 25, 30]
plt.plot(x,y,marker="o",markerfacecolor='Black', markersize=10)
plt.show()
# we can only put markers on scatter plot and line plot
```

#### Labels, Titles , gridline , Legends

```python
x= [1, 2, 3, 4]
y1= [10, 20, 25, 30]
y2= [5,15,20,25]
plt.plot(x,y1,label='Line 1') # Legends
plt.plot(x,y2,label='Line 2') # Legends
plt.title("Graph for Lables,Titles and Grids")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.grid(True) # Gridlines
```

#### Subplots

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 15, 20, 25]

plt.subplot(1, 2, 1)  # (rows, columns, panel number)
plt.plot(x, y1)
plt.title('Plot 1')

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title('Plot 2')
```

#### Arranging Plots with subplots and subplot2grid

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 15, 20, 25]

fig = plt.figure()

ax1 = plt.subplot2grid((2, 2), (0, 0))
ax2 = plt.subplot2grid((2, 2), (0, 1))
ax3 = plt.subplot2grid((2, 2), (1, 0), colspan=2)

ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y1)
ax3.plot(x, y2)

plt.show()
```

#### Sharing Axes

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 15, 20, 25]

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.plot(x, y1)
ax2.plot(x, y2)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data with a positive correlation
x = np.linspace(0, 20, 100)
y = 2 * x + 1 + np.random.normal(0, 2, 100)

plt.scatter(x, y,color='y',s=20,alpha=0.9) # alpha shows transparency
plt.title('+ve Correlation between Speed and Fuel consumption')
plt.xlabel('Speed (km/h)')
plt.ylabel('Fuel Consumption (in ml)')
plt.show()
```

#### 3.Bar plot

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 40, 30]

# Sort categories and values together in descending order
sorted_data = sorted(zip(values, categories), reverse=True)
sorted_values, sorted_categories = zip(*sorted_data)

plt.bar(sorted_categories, sorted_values)
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

#### Plotting Data from Pandas DataFrames

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'x': [1, 2, 3, 4], 'y': [10, 20, 25, 30]}
df = pd.DataFrame(data)
df.plot(x='x', y='y', kind='line', title='Plotting Data from Pandas DataFrame')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

dates = pd.date_range('20230101', periods=6)
data = {'date': dates, 'value': [1, 3, 5, 2, 4, 6]}
df = pd.DataFrame(data)
df.plot(x='date', y='value', kind='line', title='Date and Time Plotting')
plt.show()
```

#### Saving and Exporting Plots

# Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics.

### Relational Plots

```python
# Scatter plot
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.show()
```

```python
# Line plot
sns.lineplot(x='size', y='total_bill', data=df)
plt.show()
```

```python
# Barplot
sns.barplot(x='day',y='total_bill',data=df)
plt.show()
# the vertical line in each column shows confidence interval
```

```python
# Box plot
sns.boxplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
# Histogram
sns.histplot(df['total_bill'], bins=10, kde=True) # kde(kernal density estimate)--> represents the prob. density func.
plt.show()
```

### Matrix plots

```python
# Heatmap
numeric_df = df.select_dtypes(include=['float','int'])
sns.heatmap(numeric_df.corr(), annot=True)
plt.show()
```

```python
df = pd.DataFrame({
    'Age': [61, 1, 10, 1, 15, 26, 25, 22, 24],
    'Weight': [78.0, 11.0, 50.0, 13.0, 18.0, 86.6, 75.0, 80.0, 40.0],
    'Height': [1.75, 0.6, 1.2, 0.7, 1.0, 1.8, 1.7, 1.8, 1.5],
    'Salary': [70000, 15000, 25000, 20000, 30000, 90000, 85000, 82000, 45000]
})

sns.heatmap(df.corr(), annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap")
plt.show()
```
