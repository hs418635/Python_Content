**Numpy** is used with packages like SciPy(Scientific Python) and Mat-Plotlib(Plotting library) and this makes it capable to replace MATLAB.

### Indexing and slicing

Slicing Syntax
1) 1D Array: array[start:end]
2) 2D Array: array[row_start:row_end, col_start:col_end]
3) 3D Array: array[dim3_start:dim3_end, dim2_start:dim2_end, dim1_start:dim1_end]

```python
# 1D array
np.random.seed(4)
arr = np.random.randint(1,50,12)
print(arr)
print("indexing :",arr[1]) # indexing
print("slicing :",arr[0:5:2]) # step size 2 from front
```

### Array Joining and Splitting

1) np.cancatenate
2) np.vstack
3) np.hstack

##### Joining

```python
# concatenate , syntax --> np.concatenate((array1, array2, ...), axis=0)
import numpy as np
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])
# concatenate along axis 0 (rows)
result = np.concatenate((arr1,arr2),axis=0)
print(result)
```

```python
# concatenate along axis 1 (column)
result = np.concatenate((arr1,arr2),axis=1)
print(result)
```

```python
# vertical stack
result = np.vstack((arr1,arr2))
print(result)
```

```python
# horizontal stack
result = np.hstack((arr1,arr2))
print(result)
```

##### Splitting

```python
arr = np.array([1, 2, 3, 4, 5, 6])

# Split into 3 equal parts
result = np.split(arr, 3)
print(result)
```

```python
arr2 = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])

# Split into 2 parts along axis 1
result = np.split(arr2, 2, axis=1)
print(result)
```

```python
# Split at specific indices
result = np.split(arr2, [1, 3], axis=1)
print(result)
```

### Array searching and sorting

##### Array searching
1) numpy.where or using boolean indexing
2) finding the index of min and max value using --> argmin , argmax
3) finding element based on multiple conditions
4) np.searchsorted ---> is used to find indices where elements should be inserted to maintain order.

```python
# uisng where
import numpy as np
arr = np.array([10,20,30,40,50])
# finding the indices where value is 30
indices = np.where(arr==30)
print(indices)
```

```python
# using boolean indexing
arr[arr>25]
```

```python
# finding maximum and minimun values
print("index of min number :",np.argmin(arr))
print("index of max number :",np.argmax(arr))
```

```python
# finding element based on multiple conditions
arr[(arr>20) & (arr<50)]
```

```python
# 2d array
arr = np.arange(1,10).reshape(3,3)
arr
```

```python
indices = np.where(arr==6)
print(indices)
```

```python
# using boolean indexing
arr[arr>5]
```

```python
# using searchsorted()
arr = np.array([1,11,2,10,3,5,7,8,9])
arr
```

```python
np.sort(arr)
```

```python
np.searchsorted(arr,6)
```
