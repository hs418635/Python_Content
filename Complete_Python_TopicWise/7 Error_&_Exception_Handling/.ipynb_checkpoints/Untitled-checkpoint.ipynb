```python
try:
    a=int(input(' enter a number '))
    b=int(input(' enter a number '))
    print(a/b)
except ZeroDivisionError :
    print("please enter other than zero")
except ValueError :
    print("please enter integer only")
```

```python
s=int(input('enter the age of son'))
f=int(input('enter the age of father'))

try:
    assert s<=f,"son age can not be greater than father age"
except AssertionError as e:
    print(e)
else:
    print("both the ages are correct , proceeding with the further logic")
```
