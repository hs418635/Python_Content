# pickling

### Python pickle module is used for serializing and de-serializing a Python object structure. Any object in Python can be pickled so that it can be saved on disk.

##### Serializing ---> means to store object
#### deserializing ---> means to read




![Alt text](image.png)

```python
import pickle

l = [10,20,30,40]

file = open("writedata.txt",'wb') #wb ---> write binary

pickle.dump(l,file)

file.close()
```

```python
import pickle
file = open("wirtedata.txt","rb") #rb ---> read binary
l = pickle.load(file)
print(l)
```
