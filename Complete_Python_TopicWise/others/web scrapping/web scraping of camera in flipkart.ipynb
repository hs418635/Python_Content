```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import logging
```

## How to do Scrapping

```python
flipcart_url = "https://www.flipkart.com/search?q=" + "camera"
flipcart_url
```

```python
urlclient = urlopen(flipcart_url)
```

```python
flipcart_page = urlclient.read()
```

```python
# now we will use beautifulsoup to see the complete text in littlebit readable formate
flipcart_html = bs(flipcart_page , 'html.parser')
# it is now somehow readable and little bit it is going to beautify
```

```python
flipcart_html
```

```python
big_box = flipcart_html.find_all("div" , {"class":"_1AtVbE col-12-12"})
# it is going to return all the big box data wherever class id is going to match
```

```python
len(big_box)
```

```python
del big_box[0:3]
```

```python
productlink = "https://www.flipkart.com" + big_box[0].div.div.div.a['href']
```

```python
for i in big_box:
    print("https://www.flipkart.com" + i.div.div.div.a['href'])
#fetching one by one by one tv's by code only
```

```python
productlink
```

```python
product_req = requests.get(productlink)
```

```python
product_html = bs(product_req.text, 'html.parser')
```

```python
product_html
```

```python
comment_box = product_html.find_all("div", {"class":"_16PBlm"})
comment_box
```

```python
len(comment_box) # there are 10 comments in that tv product
```
