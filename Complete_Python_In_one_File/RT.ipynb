```python
import pdfplumber

class PdfHandling:

    def __init__(self):
        self.text = ""

    def read_pdf(self, filename):
        try:
            with pdfplumber.open(filename) as pdf:
                for page in pdf.pages:
                    text = page.extract_text()
                    if text:
                        self.text += text + "\n"
            return self.text if self.text.strip() else "No text found in PDF."
        except FileNotFoundError as e:
            print(f"Error: {e}")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")
        else:
            print("Program successfully ran")

# Usage
file1 = PdfHandling()
result = file1.read_pdf("mypdf.pdf")
print(result)
```
