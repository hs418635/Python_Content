### Merging DataFrames

```python
df = pd.DataFrame({
    "Name":["Ashwini","Anmol","virat","rohit",'bhuvi'],
    "Employee_no" : [12512,451254,4512,6987,2563],
    "Income":[30000,40000,2000,30000,5000],
    "Age":[21,23,24,40,21]
})
df
```

```python
df1 = pd.DataFrame({
    "Name":["sunny","rohit","dhoni","hardik",'jasprit'],
    "Employee_no" : [12512,451254,4512,6987,2563],
    "Income":[30000,40000,2000,30000,5000],
    "Age":[21,23,24,40,21]
})
df1
```

```python
## using concat()
pd.concat([df,df1])
```

```python
df2 = pd.concat([df,df1],ignore_index=True)
df2
```
