### transform() operation

```python
import pandas as pd
```

```python
import pandas as pd
df = pd.DataFrame({"Name":["aswin","anmol","rohit","ravindra","jadeja","Michael","Sarath","neelesh","Chris"],
                          "Department":["IT","Human Resources","Finance","Supply Chain","Finance","Marketing","IT","Human Resources","Law"],
                          "Income":[4800,5200,6600,5700,7200,8400,7700,4200,9400],
                          "Age":[24,28,33,41,22,46,31,27,39],
                          "Experience":[2,5,9,17,1,24,10,6,13]})
df
```

```python
def salary_increase(x):
    return x+1000
```

```python
df['Income']
```

```python
df['Income'].transform(salary_increase)
```

```python
df['Income'].transform(lambda x: x+1000)
```

```python
df['Income'].transform(lambda x: (x-x.mean())/x.std()) # standizing
```

```python
df[["Income","Age"]].transform("sqrt")
```

### Apply function 

It applies a specific funciton to all values of a dataframe

```python
df
```

```python
df["Name"].apply(len)
```

```python
def increase_age(x):
    return x+5
```

```python
df['Age'].apply(increase_age)
```

```python
import numpy as np
df.apply(np.sum)
```

```python
df[["Income"]].apply(np.sum)
```
