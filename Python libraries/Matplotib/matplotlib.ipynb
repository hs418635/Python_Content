```python
# line chart
x = [1,2,3,4]
y = [10,25,30,40]
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [10,25,30,40]
plt.plot(x,y,color='black',linestyle='dotted',linewidth=1) # dashed or dotted
plt.show()
```

```python
# markes --> o , v , ^
x = [1,2,3,4]
y = [10,25,30,40]
plt.plot(x,y,marker='o',markerfacecolor='black',markersize=10)
plt.show()
```

```python
# labels , title
```

```python
x = [1,2,3,4]
y = [10,25,30,40]
plt.plot(x,y,marker='o',markerfacecolor='black',markersize=10)
plt.xlabel('Hours study')
plt.ylabel('Marks gain')
plt.title('Marksheet')
plt.grid(True)
plt.show()
```

### subplots

```python
x = [1,2,3,4]
y1 = [10,25,30,40]
y2 = [5,15,25,30]

plt.subplot(1,2,1) # subplot(row,column,panel numbe)
plt.plot(x,y1)
plt.title('Plot 1')

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('Plot 2')

plt.show()
```

```python
# scatter plot
import numpy as np
a = np.random.randint(1,10,12)
b = np.random.randint(20,50,12)
plt.scatter(a,b)
plt.show()
```

```python
x = np.linspace(0,30,100)
y = 2*x + 1 + np.random.normal(0,2,100)
plt.scatter(x,y,color='y',s=100,alpha=0.5) # s --> size of datapoints
plt.xlabel('Speed (in kmph)')
plt.ylabel("Fuel consumption in ml")
plt.title("+ve correlation between fuel consumption and speed")
plt.show()

print(x)
print(y)
```

```python
x = np.linspace(0,30,100)
y = -2*x + 100 + np.random.normal(0,2,100)
plt.scatter(x,y,color='y',s=100,alpha=0.5) # s --> size of datapoints
plt.xlabel('Excercise')
plt.ylabel("Weight")
plt.title("-ve correlation")
plt.show()
```

### bar chart

```python
cats = ['A','B','C','D']
values = [10,20,40,30]

sorted_data = sorted(zip(values,cats),reverse=False)
sorted_values,sorted_cats = zip(*sorted_data)

plt.bar(sorted_cats,sorted_values)
plt.show()
```

```python
cats = ['A','B','C','D']
values = [10,20,40,30]

sorted_data = sorted(zip(values,cats),reverse=False)
sorted_values,sorted_cats = zip(*sorted_data)

plt.barh(sorted_cats,sorted_values,color=['r','g','y','blue'],alpha=0.5)
plt.show()
```

```python
# boxplot
data = np.random.normal(0,1,100)
plt.boxplot(data)
plt.show()
```

```python
# pie chart
cats = ['A','B','C','D']
values = [10,20,40,20]
plt.pie(values,labels=cats,autopct='%1.1f%%',counterclock=False) # anticlockwise
plt.show()
```

```python
# histogramm
data = [20,30,30,30,30,40,40,50,50,50,50]
bins = 4
plt.hist(data,bins,facecolor='blue',alpha=0.5)
plt.show()
```

```python
# plotting charts through dataframe
import pandas as pd
data = {'x':[1,2,3,4],'y':[10,20,30,40]}
df = pd.DataFrame(data)
df.plot(x='x',y='y',kind='pie')
plt.show()
```

```python
# saving charts
import pandas as pd
data = {'x':[1,2,3,4],'y':[10,20,30,40]}
df = pd.DataFrame(data)
df.plot(x='x',y='y',kind='pie')
plt.savefig("mypie.png")
plt.savefig("mypie.pdf")
plt.show()
```

```python
# correlation matrix
data = pd.DataFrame({
    'Age': [61, 1, 10, 1, 15, 26, 25, 22, 24],
    'Weight': [78.0, 11.0, 50.0, 13.0, 18.0, 86.6, 75.0, 80.0, 40.0],
    'Height': [1.75, 0.6, 1.2, 0.7, 1.0, 1.8, 1.7, 1.8, 1.5],
    'Salary': [70000, 15000, 25000, 20000, 30000, 90000, 85000, 82000, 45000]
})

sns.heatmap(data.corr(), annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap")
plt.show()
```
