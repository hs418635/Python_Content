### What is matplot lib

Matplotlib is a Python library for data visualization. It allows users to create plots, histograms, bar charts, scatter plots, etc., with just a few lines of code. Matplotlib is also used to create interactive visualizations that can be zoomed, panned, and updated.

```python
# linear (line) , scatter, bar , hist , box , pie
```

### 1.Line plot

```python
import numpy as np

a = np.array([1,2,3])
b = np.array([2,3,4])

plt.plot(a,b)
```

### Markers

```python
# v ---> downward triangle , * , o---> cicle , ^ ---> for upward traingle , d--> diamond
a = np.array([1,12,3,23])
plt.plot(a, marker='d')
```

```python
# dotted and dashed lines
a = np.array([1,12,3,23])
plt.plot(a,linestyle='dashed',marker='o')
```

```python
### Labels , titles , and grid

a = np.array([1,2,3])
b = np.array([2,3,4])

plt.plot(a,b)
plt.title("Graph for labels, titles and grid")
plt.xlabel("study hours")
plt.ylabel("marks")
plt.grid() # grid
```

### subplot

```python
# subplot
#plot1
a = np.array([1,2,3])
b = np.array([2,3,4])
plt.subplot(1,2,1) #(whole_section,number_of_division,place_to_put_your_graph)
plt.plot(a,b)


#plot2
x = np.array([4,2,7])
y = np.array([3,8,6])
plt.subplot(1,2,2) #(whole_section,number_of_division,place_to_put_your_graph)
plt.plot(x,y)
```

## 3.bar plot

```python
import numpy as np
import matplotlib.pyplot as plt
objects = ('p','y','t','h','o','n')
y_pos = np.arange(len(objects))
l = [10,8,6,4,2,1]
sorted_l = np.sort(l)
#plt.bar(y_pos,l) # unsorted
plt.bar(y_pos,sorted_l) # sorted
```

```python
# horizontal bar ---> plt.barh
objects = ('p','y','t','h','o','n')
y_pos = np.arange(len(objects))
l = [10,8,6,4,2,1]
sorted_l = np.sort(l)
#plt.bar(y_pos,l) # unsorted
plt.barh(y_pos,sorted_l) # sorted
```

## colors

```python
# horizontal bar ---> plt.barh
c = ['r','y','g','b','c','m'] # color
objects = ('p','y','t','h','o','n')
y_pos = np.arange(len(objects))
l = [10,8,6,4,2,1]
sorted_l = np.sort(l)
#plt.bar(y_pos,l) # unsorted
plt.barh(y_pos,sorted_l,color=c) # sorted
```

```python
x = [10,10,10,10,10,15,15,15,15,20,20,20,25,30,30,30,30,35,40,40,40]
num_bins = 7
n,bins,patches = plt.hist(x, num_bins, facecolor = 'blue' , alpha=0.5)
plt.xlabel("number")
plt.ylabel("frequency")
plt.title("histogram")
# plt.grid()
```
